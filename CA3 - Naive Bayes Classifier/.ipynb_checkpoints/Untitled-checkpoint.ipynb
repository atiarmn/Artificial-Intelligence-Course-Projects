<p>The goal of this computer assignment is to classify comments by two classes(recommended and not recommended) using Naive Bayes classifier and bag of words model.</p>
<h3>Data Processing</h3>
<p>Before modeling the problem by Naive Bayes model, comments sould be processed.<br>Comments and their titles are combined and they are normalized and tokenized using hazm library.
Next step is to delete some repetitive words from the comments. Hazm library has a set of repetitive words named Stopwords_list() for this issue.Lemmatizer - a class from hazm library - has a function named lemmatize to find infinitive of a verb. Stemmer - a class from hazm library - has a function named stem that delete the extra letters attached to the words. Using these functions, words can be simplified and identified and matched more easily.<br></p>
</p>Data_processing function is processing comments and if it's used for the train file it will return 1 list and 2 dictionarys: bag of words , recommended words and not recommended words and if it's used for the test file it will just return bag of words list. Bag of words contains every normalized, lemmatized and non repetitive words in the file. Recommended words and not recommended words are sublists of bag of words and they contain the words that were used in recommended comments and not recommended comments.</p> 

```python
import pandas as pd
from hazm import *
from math import log2

def data_processing(df,train):
    bag_of_words = []
    sentence = []
    recommended_words = {}
    not_recommended_words = {}

    df['comment'] = df['comment'] + ' ' +df['title']
    
    for index, row in df.iterrows():
        row['comment'] = normalizer.normalize(row['comment'])
        row['comment'] = word_tokenize(row['comment'])
        sentence = row['comment']
        j = 0
        while (j in range(len(sentence))):
            if(sentence[j] in stopwords_list()):
                del(sentence[j])
                j -= 1
            j += 1
        
        for word in sentence:
            word = lemmatizer.lemmatize(word)
            #word = stemmer.stem(word)
            if(train):    
                if(row['recommend'] == 'recommended'):
                    if(word in recommended_words):
                        recommended_words[word] += 1
                    else:
                        recommended_words[word] = 1
                else:
                    if(word in not_recommended_words):
                        not_recommended_words[word] += 1
                    else:
                        not_recommended_words[word] = 1
                        
            bag_of_words.append(word)
    if(not train):
        return bag_of_words
    return bag_of_words , recommended_words , not_recommended_words

normalizer = Normalizer()
lemmatizer = Lemmatizer()
stemmer = Stemmer()

train_df = pd.read_csv("comment_train.csv")
train_bag_of_words , train_recommended_words , train_not_recommended_words = data_processing(train_df,train = True)
test_df = pd.read_csv("comment_test.csv")
test_bag_of_words = data_processing(test_df,train = False)
```

<h3>Naive Bayes Classifier</h3>
<p>The problem is to use naive bayes classifier and detect recommended and not recommended comments. The first step is to train the system by a comment_train.csv file that has a lot of comments and it shows that the comment meant that the user recommended the product or not. Thus, the words of the comments that were recommended and not recommended are returned by the data processing function and shown as train recommended_words and train_not_recommended_words.<br>
Detecting recommended and not recimmended comments using naive bayes means calculating P(recommended|words of the comment) and P(not recommended|words of the comment). Bayes Theorem shows that for events A and B, provided that P(B) ≠ 0:<br> P(A|B) = [P(B|A)×P(A)]/P(B) <br>
This means that posterior probabilities are proportional to the numerator and P(A|B) ∝ P(A,B). Thus, calculating P(recommended,words of the comment) and P(not recommended,words of the comment) and get the maximum of them will give the result.<br></p>
<h5>Evidence:</h5><p>The evidences of this problem are the words that are in the train file and their probability of the recommended or not recommended of the comment that contains them can be calculated.</p>
<h5>Liklihood:</h5><p>The likelihood is the probability of being recommended or not recommended of the comment. This probability can be calculated by words of the comment being in the "train recommended words" or "train not recommended words" dictionaries.</p>
<h5>Prior:</h5><p>The prior probabilities are the probabilities that can be calculated by the information given by the train file.Like probability of recommended comments and not recommended comments.</p>
<h5>Posterior:</h5><p>The posterior probability is what the problem wants - probability of a comment which has word x, being in a class ( recommended or not recommended ).</p>

<p>The conditional_prob function calculates P(recommended,words of the comment) and P(not recommended,words of the comment) using train recommended words and train not recommended words. <br>
Sometimes there is a word that is in the "train recommended words" dictionary but not in the "train not recommended words" dictionary.Thus, based on naive bayes P(word|not recommended) will be 0 and so the product of the P(not recommended,words of the comment) will be 0. So the system will take the comment with that word definitely in the recommended class which can be not correct. "Additive Smoothing" is a technique used to avoid this problem. Additive smoothing will add one to the number of repetition of every word and make the probability (x + 1)/(N + d) instead of x/N.<br>Using logarithm and sum instead of multiplication helps avoiding very small and fractional numbers.</p>
<p>The calculate_recommended_chance is a function that gets the logarithm of probability of the comment being recommended or not, find the maximum and decides whether the comment is more probable to be in the recommended or not recommended class.</p>

```python
def conditional_prob_additive(word,alpha,d):
    recommended_count = 0
    not_recommended_count = 0
    if(word in train_recommended_words):
        recommended_count += train_recommended_words[word]
    if(word in train_not_recommended_words):    
        not_recommended_count += train_not_recommended_words[word]
    recommended_prob = (recommended_count+alpha) / (recommend_count[0]+(len(train_not_recommended_words)*alpha))
    not_recommended_prob = (not_recommended_count+alpha) / (recommend_count[1]+(len(train_not_recommended_words)*alpha))
    return log2(recommended_prob) , log2(not_recommended_prob)

def conditional_prob(word):
    recommended_count = 0
    not_recommended_count = 0
    if(word in train_recommended_words):
        recommended_count += train_recommended_words[word]
    if(word in train_not_recommended_words):    
        not_recommended_count += train_not_recommended_words[word]
    recommended_prob = (recommended_count) / (recommend_count[0])
    not_recommended_prob = (not_recommended_count) / (recommend_count[1])
    if(recommended_prob == 0 and not_recommended_prob == 0):
        return 0 , 0
    elif(recommended_prob == 0 and not_recommended_prob != 0):
        return 0 , log2(not_recommended_prob)
    elif(recommended_prob != 0 and not_recommended_prob == 0):
        return log2(recommended_prob) , 0
    else:
        return log2(recommended_prob) , log2(not_recommended_prob)

def calculate_recommended_chance(comment,train_bag_of_words_length,recommended_prob,not_recommended_prob):
    recommended_product = 0
    not_recommended_product = 0
    word_recommended_prob = 0
    word_not_recommended_prob = 0
    for word in comment:
        word_recommended_prob , word_not_recommended_prob = conditional_prob_additive(word,alpha = 1,d = train_bag_of_words_length)
        #word_recommended_prob , word_not_recommended_prob = conditional_prob(word)
        #if(word_recommended_prob == 0 or word_not_recommended_prob == 0):
         #   break
        recommended_product += word_recommended_prob
        not_recommended_product += word_not_recommended_prob
    #if(word_recommended_prob == 0):
     #   return "not_recommended"
    #elif(word_not_recommended_prob == 0):
     #   return "recommended"

    recommended_result = recommended_prob + recommended_product
    not_recommended_result = not_recommended_prob + not_recommended_product
    if(not_recommended_result > recommended_result):
        return "not_recommended"
    else:
        return "recommended"

recommend_count = train_df['recommend'].value_counts() 
recommended_prob = log2(recommend_count[0] / len(train_df))
not_recommended_prob = log2(recommend_count[1] / len(train_df))
train_bag_of_words_length = len(train_bag_of_words)

test_df['My guess'] = test_df.apply(lambda row : calculate_recommended_chance(row['comment'],train_bag_of_words_length,recommended_prob,not_recommended_prob), axis = 1)
```

<h3>Evaluation</h3>
<p>Here are four evaluation criteria:Accuracy,Precision,Recall and F1.</p>
<h5>Accuracy:</h5><p>Accuracy is Correct detected comments divided by Total comments.</p>
<h5>Precision:</h5><p>Precision is the number of correctly identified recommended comments divided by the number of all detected recommended comments .<br>This criterion is not valid on it's own because it just focus on the accuracy of the comments that are detected as recommended comments.For example if 10% of the recommended comments were ditected and the rest of them was ditected as not recommended comment, the precision will be 1 but the result was not correct.</p>

<h5>Recall:</h5><p>Recall is the number of correctly identified recommended comments divided by the number of all samples that should have been identified as recommended comments.<br>This criterion is not valid on it's own because it doesn't estimate any "not recommended comment". For example if the number of all samples was 1400 and the number of all samples that should have been identified as recommended comments was 700 and all the comments was considered as recommended, the recall was 1 but the result that was not correct. </p>
<h5>F1:</h5><p>F1 is a measure of a test's accuracy and it's calculated from the precision and recall of the test.F1 is the harmonic mean of precision and recall and it's needed when seeking a balance between precision and recall is needed.</p>



```python
guess_count = test_df['My guess'].value_counts().sort_index() 
recommend_count = test_df['recommend'].value_counts().sort_index()

accuracy = len(test_df[test_df['recommend'] == test_df['My guess']]) / len(test_df)
print(accuracy)

correct_recommended = len(test_df[(test_df['recommend'] == test_df['My guess']) & (test_df['My guess']=='recommended')])
precision = correct_recommended / guess_count[1]
print(precision)

recall = correct_recommended / recommend_count[1]
print(recall)

f1 = 2 * ((precision*recall)/(precision + recall))
print(f1)
```
