<h3>Preprocessing</h3>
<h5>Handling Missing Data</h5>
<p>A number of alternative ways of handling the missing data has been developed. Dropping a column which has a lot of missing data is one of the ways of handling missing data but it is not a good way for every column in the dataframe because suppose we have a column which have only 0.05% missing data, we can not drop this data because of this amount of missing data because it may have important information. But if there is a column with 20% missing data, it can be dropped.<br>Another way to handle missing data is to use the mean value of a variable in place of the missing data value for that same variable. This can be used for only numeric features.<br>Regression imputation is another way of handling missing data which the existing variables are used to make a prediction, and then the predicted value is substituted as if an actual obtained value. This may be better way compare to replace missing data with mean value of the column because when we use every available data to predict the missing data, that prediction can be very accurate and very close to the actual value of that data.<br> For non numeric featuresattribiutes it is hard to use the last two ways mentioned above. One way to handle missing data in these columns is to replace them with the most repetitive value in the column.</p>
<p>In last section we saw that "LotFrontage" and "GarageYrBlt" has the most missing data among all the columns. Thus dropping them won't be a problem.<p>

```python
def handle_missing_data(df):
    values = {'Alley': 'NOTEX', 'BsmtQual':'NOTEX', 'BsmtCond':'NOTEX', 'BsmtExposure':'NOTEX', 'BsmtFinType1':'NOTEX', 
          'BsmtFinType2':'NOTEX', 'FireplaceQu':'NOTEX', 'GarageType': 'NOTEX', 'GarageFinish': 'NOTEX',
          'GarageQual': 'NOTEX', 'GarageCond': 'NOTEX', 'PoolQC': 'NOTEX', 'Fence': 'NOTEX', 'MiscFeature': 'NOTEX'}
    df.fillna(value = values, inplace = True)
    df.drop(columns = ['LotFrontage','GarageYrBlt'],inplace = True)
    for (columnName, columnData) in df.iteritems():
        if(df[columnName].isna().any()):
            if(df[columnName].dtypes == 'object'):
                val_count = df[columnName].value_counts()
                df[columnName].fillna(val_count.idxmax(),inplace = True)
            else:
                df[columnName].fillna(df[columnName].mean(),inplace = True)
    
```

<h5>Delete Unnecessary Features </h5>
<p>In visualisation and EDA section we saw the correlation heatmap. The features that has the minimum correlation with "SalePrice" - the target value of this project - can be easily deleted.</p>

```python
def delete_unnecessary_features(df):
    limit = 0.2
    for col in df.columns:
        if df.corr()['SalePrice'][col] < limit:
            if col == 'Id':
                continue
            df.drop(col, inplace = True, axis = 1)
```

<h5>Normalizing and Standardizing</h5>
<p>Data normalization is used to make model training less sensitive to the scale of features and makes the data better conditioned for convergence. In general, the goal of normalization is to change the values of numeric columns in the dataset to a common scale, without distorting differences in the ranges of values.</p>
<p>Standardizing the features around the center and 0 with a standard deviation of 1 is important when we compare measurements that have different units. Variables that are measured at different scales do not contribute equally to the analysis and might end up creating a bais.</p>

```python
from sklearn import preprocessing
from pandas.plotting import andrews_curves

def normalizing_and_standardizing(df):
    num_cols = []

    for col in df.columns[df.dtypes.apply(lambda c: np.issubdtype(c, np.number))]:
        if ((col != 'Id') and (col != 'SalePrice')):
            num_cols.append(col)
    
    normalized_df = df.copy()
    normalized_df[num_cols] = preprocessing.normalize(normalized_df[num_cols])


    standardiser = preprocessing.StandardScaler()
    standardised_df = df.copy()
    standardised_df[num_cols] = standardiser.fit_transform(standardised_df[num_cols])
    return normalized_df, standardised_df
```

<h5>Categorical Featue Encoding</h5>
<p>There are 3 different types of categorical data:</p>
<li>Binary</li>
<ol>Binary features are those with only two possible values. The values of these features can easly replace with two numbers (0,1) using replace() function.</ol>
<li>Ordinal </li>
<ol>Ordinal features are those with some order associated with them. The values of these features also can replace with numbers (integer encoding) which the order of them will show the order of the values of the feature. Map() will map the numbers to the labels.</ol>
<li>Nominal</li>
<ol>Nominal features are categorical features that have no numerical importance. One-hot encoding can be used to encode these features. In this encoding, a new binary column is created for every value that the feature can get and for each of them the column corresponding to that value gets 1 and the others get 0. </ol>

```python
def categorical_encoding(df):
    ordinal_dict = {"Ex": 5, "Gd": 4, "TA": 3, "Fa": 2, "Po": 1, "NOTEX": 0,
               "MnWw":4, "GdWo":3, "MnPrv":2, "GdPrv":1,
               "Fin":3, "RFn":2, "Unf":1,
               "Y":1, "N":0,
               "GLQ":6, "ALQ":5, "BLQ":4, "Rec":3, "LwQ":2, "Unf":1,
               "Av":3, "Mn":2, "No":1,
               "Gtl":3, "Mod":2, "Sev":1,
               "AllPub":4, "NoSewr":3, "NoSeWa":2, "ELO":1, 
               "Reg":4, "IR1":3, "IR2":2, "IR3":1,
               "Grvl":2, "Pave":1}
    df['Street'] = df.Street.map(ordinal_dict)
    df['Alley'] = df.Alley.map(ordinal_dict)
    df['LotShape'] = df.LotShape.map(ordinal_dict)
    df['Utilities'] = df.Utilities.map(ordinal_dict)
    df['LandSlope'] = df.LandSlope.map(ordinal_dict)
    df['ExterQual'] = df.ExterQual.map(ordinal_dict)
    df['ExterCond'] = df.ExterCond.map(ordinal_dict)
    df['BsmtQual'] = df.BsmtQual.map(ordinal_dict)
    df['BsmtCond'] = df.BsmtCond.map(ordinal_dict)
    df['BsmtExposure'] = df.BsmtExposure.map(ordinal_dict)
    df['BsmtFinType1'] = df.BsmtFinType1.map(ordinal_dict)
    df['BsmtFinType2'] = df.BsmtFinType2.map(ordinal_dict)
    df['HeatingQC'] = df.HeatingQC.map(ordinal_dict)
    df['CentralAir'] = df.CentralAir.map(ordinal_dict)
    df['KitchenQual'] = df.KitchenQual.map(ordinal_dict)
    df['FireplaceQu'] = df.FireplaceQu.map(ordinal_dict)
    df['GarageFinish'] = df.GarageFinish.map(ordinal_dict)
    df['GarageQual'] = df.GarageQual.map(ordinal_dict)
    df['GarageCond'] = df.GarageCond.map(ordinal_dict)
    df['PoolQC'] = df.PoolQC.map(ordinal_dict)
    df['Fence'] = df.Fence.map(ordinal_dict)

    nominal_features = []
    for col in df.columns:
        if df[col].dtypes == 'object':
            nominal_features.append(col)
    df = pd.get_dummies(df, columns = nominal_features, drop_first = True)
    return df

```

<h5>Cross Validation</h5>
<p>Once training the model is done, it should be evaluated to assume that it is going to work well on data that it has not seen before.<br>
Testing any machine learning model on some unseen data is a way to evaluate the performance of it. Based on the models performance on unseen data we can say weather our model is Under-fitting/Over-fitting/Well generalized. Cross validation (CV) is one of the technique used to test the effectiveness of a machine learning models, it is also a re-sampling procedure used to evaluate a model if we have a limited data. Cross validation is one of the technique used to test the effectiveness of a machine learning models, it is also a re-sampling procedure used to evaluate a model if we have a limited data.</p>
<p>Common techniques used for cross validation:</p>
<li>Leave p-out cross-validation:</li>
<ol>This technique involves using p-observation as validation data, and remaining data is used to train the model.</ol>
<li>Leave-one-out cross-validation:</li>
<ol>For a model having n samples, first sample is selected for validation and the rest (n-1) samples are used to train the model. For next iteration, second sample is selected for validation and rest to train the model. And this process continues in the same way.</ol>
<li>Holdout cross-validation:</li>
<ol>In this technique, dataset is randomly split into training and validation data. You should know that the more data is used to train the model, the better the model is.
With this approach there is a possibility of high bias if we have limited data, because we would miss some information about the data which we have not used for training.</ol>
<li>K-fold cross-validation:</li>
<ol>The original dataset is equally partitioned into k subsets. For each iteration, one of the k subsets or folds is selected as validation data and the remaining subsets are selected as training data. The process is repeated for k times until each group is once trated as validation data.</ol>
<li>Repeated random subsampling validation:</li>
<ol>In this technique, dataset randomly split into training and validation subsets and this random subsampling validation is repeated in a row. One of the disadvantages of this method is that some samples may not be selected for either training or validation.</ol>

<p>Here, holdout cross-validation is used by train_test_split() function.</p>

```python
from sklearn.model_selection import train_test_split
def split(df):
    train, validation = train_test_split(df, test_size = 0.3, random_state = 42, shuffle = True)
    return train, validation
```

<h3>Overfitting and Underfitting</h3>
<p>Formally, overfitting referes to the situation where a model learns the data but also the noise that is part of training data to the extent that it negatively impacts the performance of the model on new unseen data.<br>A model that underfits the data will have poor performance on the training data.</p>
<h3>Model Training, Evaluation and Hyperparameter Tuning</h3>
<p>Here are the implementations of three models: K-Nearest-Neighbers, Decision Tree and Linear Regression. <br> K-Nearest-Neighbers and Decision Tree has a several hyperpatameters. <h5>K-Nearest-Neighbers</h5>
<p>Based on the diagrams bellow, we can say that the proper neighber number is about three or four. Additionally, the model is underfitting when neighber number is about 12. </p>

```python
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import mean_absolute_error, mean_squared_error

plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')

def KNN(df,hyper_param_grid):
    train_df, validation_df = split(df)
    train_inputs = train_df.drop(columns=['SalePrice','Id'])
    train_target = train_df['SalePrice'].values
    validation_target = validation_df['SalePrice'].values
    validation_inputs = validation_df.drop(columns=['SalePrice','Id'])
    
    validation_MAE = []
    validation_RMSE = []
    train_MAE = []
    train_RMSE = []
    for k in hyper_param_grid:
        k_neighbors = KNeighborsClassifier(n_neighbors = k)
        k_neighbors.fit(train_inputs,train_target)
        
        validation_pred = k_neighbors.predict(validation_inputs)
        train_pred = k_neighbors.predict(train_inputs)
        
        validation_MAE.append(mean_absolute_error(validation_target,validation_pred))
        validation_RMSE.append(mean_squared_error(validation_target, validation_pred, squared=False))
        train_MAE.append(mean_absolute_error(train_target,train_pred))
        train_RMSE.append(mean_squared_error(train_target, train_pred, squared=False))
    
    
    plt.subplot(1,2,1)
    plt.plot(hyper_param_grid, np.array(validation_MAE).ravel(), 'o-', label='validation')
    plt.plot(hyper_param_grid, np.array(train_MAE).ravel(), 'o-', label='train')
    plt.xlabel('Neighber')
    plt.ylabel('MAE')
    plt.title('KNN')
    plt.tight_layout()
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(hyper_param_grid, np.array(validation_RMSE).ravel(), 'o-', label='validation')
    plt.plot(hyper_param_grid, np.array(train_RMSE).ravel(), 'o-', label='train')
    plt.xlabel('Neighber')
    plt.ylabel('RMSE')
    plt.title('KNN')
    plt.tight_layout()
    plt.legend()
    
    plt.show()
    
    
df = pd.read_csv("train.csv")
handle_missing_data(df)
df = categorical_encoding(df)
normalized_df, standardized_df = normalizing_and_standardizing(df)
delete_unnecessary_features(standardized_df)
hyper_param_grid = np.arange(1, 25)
KNN(standardized_df, hyper_param_grid)
```

<h5>Decision Tree</h5>
<p>Based on the diagrams bellow, we can say that the proper maximum depth is about 13 or 14. Additionally, the model is overfitting when maximum depth is about 23 or 24 and it is underfitting when maximum depth is 1. </p>

```python
from sklearn.tree import DecisionTreeClassifier

plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')


def decision_tree(df,depth_range):
    train_df, validation_df = split(df)
    train_inputs = train_df.drop(columns=['SalePrice','Id'])
    train_target = train_df['SalePrice'].values
    validation_target = validation_df['SalePrice'].values
    validation_inputs = validation_df.drop(columns=['SalePrice','Id'])
    
    validation_MAE = []
    validation_RMSE = []
    train_MAE = []
    train_RMSE = []
    for i in depth_range:
        decision_tree_classifier= DecisionTreeClassifier(random_state = 42, max_depth=i)
        decision_tree_classifier = decision_tree_classifier.fit(train_inputs,train_target)
        validation_pred = decision_tree_classifier.predict(validation_inputs)
        train_pred = decision_tree_classifier.predict(train_inputs)
        
        validation_MAE.append(mean_absolute_error(validation_target,validation_pred))
        validation_RMSE.append(mean_squared_error(validation_target, validation_pred, squared=False))
        train_MAE.append(mean_absolute_error(train_target,train_pred))
        train_RMSE.append(mean_squared_error(train_target, train_pred, squared=False))
    
    plt.subplot(1,2,1)
    plt.plot(depth_range, np.array(validation_MAE).ravel(), 'o-', label='validation')
    plt.plot(depth_range, np.array(train_MAE).ravel(), 'o-', label='train')
    plt.xlabel('Iterations')
    plt.ylabel('MAE')
    plt.title('Decision Tree')
    plt.tight_layout()
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(depth_range, np.array(validation_RMSE).ravel(), 'o-', label='validation')
    plt.plot(depth_range, np.array(train_RMSE).ravel(), 'o-', label='train')
    plt.xlabel('Iterations')
    plt.ylabel('RMSE')
    plt.title('Decision Tree')
    plt.tight_layout()
    plt.legend()
    
    plt.show()
    
df = pd.read_csv("train.csv")
handle_missing_data(df)
df = categorical_encoding(df)
normalized_df, standardized_df = normalizing_and_standardizing(df)
delete_unnecessary_features(standardized_df)
depth_range = np.arange(1, 25)
decision_tree(standardized_df,depth_range)
```

<h3>Ensemble Methods</h3>
<p>Ensemble methods are techniques that create multiple models and then combine them to produce improved results. Ensemble methods usually produces more accurate solutions than a single model would. This has been the case in a number of machine learning competitions, where the winning solutions used ensemble methods.<br>In this project, two ensemble methods are used: Random Forest and Voting Regression</p>

```python
from sklearn.ensemble import RandomForestRegressor

def random_forest(df,estimators,max_features_range):
    train_df, validation_df = split(df)
    train_inputs = train_df.drop(columns=['SalePrice','Id'])
    train_target = train_df['SalePrice'].values
    validation_target = validation_df['SalePrice'].values
    validation_inputs = validation_df.drop(columns=['SalePrice','Id'])
    
    validation_MAE = []
    validation_RMSE = []
    train_MAE = []
    train_RMSE = []
    for i in max_features_range:
        random_forest_reg = RandomForestRegressor(n_estimators = estimators,max_features = i , random_state = 42)
        random_forest_reg.fit(train_inputs, train_target);
        validation_pred = random_forest_reg.predict(validation_inputs)
        train_pred = random_forest_reg.predict(train_inputs)
        
        validation_MAE.append(mean_absolute_error(validation_target,validation_pred))
        validation_RMSE.append(mean_squared_error(validation_target, validation_pred, squared=False))
        train_MAE.append(mean_absolute_error(train_target,train_pred))
        train_RMSE.append(mean_squared_error(train_target, train_pred, squared=False))
    
    plt.subplot(1,2,1)
    plt.plot(max_features_range, np.array(validation_MAE).ravel(), 'o-', label='validation')
    plt.plot(max_features_range, np.array(train_MAE).ravel(), 'o-', label='train')
    plt.xlabel('Max feature')
    plt.ylabel('MAE')
    plt.title('random_forest')
    plt.tight_layout()
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(max_features_range, np.array(validation_RMSE).ravel(), 'o-', label='validation')
    plt.plot(max_features_range, np.array(train_RMSE).ravel(), 'o-', label='train')
    plt.xlabel('Max feature')
    plt.ylabel('RMSE')
    plt.title('random_forest')
    plt.tight_layout()
    plt.legend()
    
    plt.show()
        
df = pd.read_csv("train.csv")
handle_missing_data(df)
df = categorical_encoding(df)
normalized_df, standardized_df = normalizing_and_standardizing(df)
delete_unnecessary_features(standardized_df)
max_features_range = np.arange(4,20)
estimators = 25
random_forest(standardized_df,estimators,max_features_range)
estimators = 100
random_forest(standardized_df,estimators,max_features_range)
```
